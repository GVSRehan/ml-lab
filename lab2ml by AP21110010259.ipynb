```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the iris dataset
iris = datasets.load_iris()

# Convert to DataFrame
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                       columns=iris['feature_names'] + ['species'])

# Replace the values with class labels
iris_df.species = np.where(iris_df.species == 0.0, 'setosa',
                           np.where(iris_df.species == 1.0, 'versicolor', 'virginica'))

# Remove spaces from column names
iris_df.columns = iris_df.columns.str.replace(' ', '')

# Describe the dataset
print(iris_df.describe())

# Check the frequency distribution for each category in the 'species' column
species_counts = iris_df['species'].value_counts()
print(species_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Plot histogram for each feature in the iris dataset
iris_df.hist()
plt.suptitle("Histogram", fontsize=16)
plt.show()

# Plot boxplot for each feature in the iris dataset
iris_df.boxplot()
plt.title("Box Plot", fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()
iris_df = sns.load_dataset('iris')

# Calculate the mean of each feature for each label class (species)
mean_by_species = iris_df.groupby(by="species").mean()

# Plot a bar chart for mean measurements
mean_by_species.plot(kind="bar")
plt.title('Class vs Measurements')
plt.ylabel('Mean Measurement (cm)')
plt.xticks(rotation=0)  # Manage the xticks rotation
plt.grid(True)
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the species information to the DataFrame
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Create a correlation matrix excluding the 'species' column
numeric_features = iris_df.select_dtypes(include=['float64', 'int64'])
corr = numeric_features.corr()

# Print the correlation matrix
print(corr)

# Plot the correlation matrix using seaborn
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Iris Dataset Features')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Create a pair plot
sns.pairplot(iris_df, hue='species')

# Use suptitle to add title to all subplots
plt.suptitle("Pair Plot", fontsize=20)

# Show the plot
plt.show()
```
